```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

# Loading Dataset

```python
d = pd.read_csv(r"C:\Users\Admin\Downloads\general_data.csv")
d.head()
```

# Dropping Irrevelant Columns

```python
d1 = d.drop(['EmployeeCount','EmployeeID','Over18','StandardHours'],axis=1)
d1.head()
```

# Checking Information of the Dataset

```python
d.info()
```

# Checking Null Values

```python
d.isna().sum()
```

# Filling Null Values with Mean

```python
d1 = d.fillna(d.mean())
d1.isna().sum()
```

# Encoding Categorial Columns

```python
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
label_encoder = preprocessing.LabelEncoder()
```

```python
d1["Attrition"] = label_encoder.fit_transform(d["Attrition"])
d1["Gender"] = label_encoder.fit_transform(d["Gender"])
d1["JobRole"] = label_encoder.fit_transform(d["JobRole"])
d1["MaritalStatus"] = label_encoder.fit_transform(d["MaritalStatus"])
d1["BusinessTravel"] = label_encoder.fit_transform(d["BusinessTravel"])
d1["Department"] = label_encoder.fit_transform(d["Department"])
d1["EducationField"] = label_encoder.fit_transform(d["EducationField"])
```

```python
d1.head()
```

```python
d1 = d1.drop(['EmployeeCount','EmployeeID', 'Over18','StandardHours'], axis=1)
d1.head()
```

```python
d1.shape
```

# Splitting the above Dataset

```python
y = d1.iloc[:, 1:2]
y.head()
```

```python
x = d1.drop(['Attrition'], axis=1)
x.head()
```

```python
x.shape
```

# Applying Decision Tree

```python
from sklearn import tree
```

```python
d1.head()
```

```python
d1.shape
```

```python
x.shape
```

# Training the model using Decision Tree Classification

```python
model = tree.DecisionTreeClassifier()
model.fit(x,y)
```

# Calculating Model Score

```python
model.score(x,y)
```

# Applying Random Forest Algorithm

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=1000, max_features=2, oob_score=True)
```

```python
rf.fit(x,y)
```

# Checking Score

```python
rf.oob_score_
```

# Finding Most Relevant Column for predicting Attrition

```python
a1 = ['Age', 'BusinessTravel', 'Department', 'DistanceFromHome','Education', 'EducationField', 'Gender', 'JobLevel', 'JobRole',
       'MaritalStatus', 'MonthlyIncome', 'NumCompaniesWorked','PercentSalaryHike', 'StockOptionLevel', 'TotalWorkingYears',
       'TrainingTimesLastYear', 'YearsAtCompany', 'YearsSinceLastPromotion','YearsWithCurrManager']
for feature,imp in zip(a1, rf.feature_importances_):
    print(feature,"\t\t :",imp);
```

# Inference 

Age, DistanceFromHome, MonthlyIncome, PercentSalaryHike, TotalWorkingYears, YearsAtCompany have value more than 0.05 ie, They are Important feature and affect the prediction of Attrition

# Spliting Dataset according to Important Columns

```python
imp = d1[['Age','DistanceFromHome','MonthlyIncome','PercentSalaryHike','TotalWorkingYears','YearsAtCompany']]
imp.head()
```

# Fitting Decision Tree

```python
model = tree.DecisionTreeClassifier(max_depth=9)
model.fit(imp,y)
```

# Creating Random Forest Tree

```python
from sklearn.tree import export_graphviz
from IPython.display import SVG
from graphviz import Source
from IPython.display import display
```

```python
graph = Source(tree.export_graphviz(model, out_file=None, 
                                    feature_names=['Age','DistanceFromHome','MonthlyIncome','PercentSalaryHike',
                                                   'TotalWorkingYears','YearsAtCompany'],
                                    class_names=['0','1'] , filled = True))
display(SVG(graph.pipe(format='svg')))
```

# Score Prediction of our Model

```python
model.score(imp,y)
```

# Fitting Random Forest to our model

```python
rf.fit(imp,y)
```

# Accuracy of our Trained Model

```python
rf.oob_score_
```

# Observations :

The accuracy score remains aproximately same in case of Random Forest while predicting on Important Columns than predicting on all Columns.

The Data is Not overfitted as in the first case of prediction using Decision Tree ( prediction of all columns : accuracy = 100%)
